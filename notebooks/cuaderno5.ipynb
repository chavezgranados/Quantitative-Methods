### **2.2 Medidas de forma**

```python
af=num2.skew()
af
```

```python
ap = 3 * (num2.mean() - num2.median()) / num2.std()
ap
```

```python
q1 = num2.quantile(0.25)
q2 = num2.quantile(0.50)
q3 = num2.quantile(0.75)

ab = (q3 + q1 - 2*q2) / (q3 - q1)
ab
```

```python
p10 = num2.quantile(0.10)
p50 = num2.quantile(0.50)
p90 = num2.quantile(0.90)

ak = (p90 + p10 - 2*p50) / (p90 - p10)
ak
```

```python
asimetria = pd.DataFrame({
    "Fisher": num2.skew(),
    "Pearson": 3 * (num2.mean() - num2.median()) / num2.std(),
    "Bowley": (num2.quantile(0.75) + num2.quantile(0.25) - 2*num2.quantile(0.50)) /
              (num2.quantile(0.75) - num2.quantile(0.25)),
    "Kelly": (num2.quantile(0.90) + num2.quantile(0.10) - 2*num2.quantile(0.50)) /
             (num2.quantile(0.90) - num2.quantile(0.10))
})

asimetria.round(3)
```

## **Anexo: datos agrupados**

Sea $x$ una variable discreta con $k$ valores diferentes,
$$\bar x=\dfrac{\sum f_ix_i}{n}, \quad \forall i=1,...,k$$
Si $x$ es una variable continua con $k$ intervalos,
$$\bar x=\dfrac{\sum f_im_i}{n}, \quad \forall i=1,...,k$$

Manual:

**Tarea**

Automático:

```python
fi = pd.cut(datos["Fare"], bins=10).value_counts().sort_index()
n = len(datos["Fare"])
intervalos = fi.index
m = [(i.left + i.right) / 2 for i in intervalos]
tabla = pd.DataFrame({
    "fi": fi
})
# Agregar marca de clase como segunda columna
tabla.insert(0, "m", m)
# Frecuencia acumulada
tabla["Fi"] = tabla["fi"].cumsum()
# Frecuencia relativa
tabla["hi"] = tabla["fi"] / n
# Frecuencia relativa acumulada
tabla["Hi"] = tabla["hi"].cumsum()
tabla
```
