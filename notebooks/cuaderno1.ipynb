```python
def leer_enaho(zip_file, tabla):

    with zipfile.ZipFile(zip_file) as z:
        archivo = next(f for f in z.namelist()
            if tabla in f)
        return pd.read_csv(
            z.open(archivo),
            sep=";",
            encoding="latin1",
            low_memory=False)
base = "./2025/Condiciones de Vida y Pobreza - ENAHO/2025/Anual - (Ene-Dic)"
```

## **2. Fusionando años**

```python
enaho = filter_catalog(catalog, survey="enaho", year_min=2023, year_max=2025, period="Anual")
download_modules(enaho, dest="./enaho", fmt="CSV")
```

```python
base1 = "/content/enaho/Condiciones de Vida y Pobreza - ENAHO"
```

```python
def leer_enaho(zip_file, tabla):
    with zipfile.ZipFile(zip_file) as z:
        archivo = next(
            f for f in z.namelist()
            if tabla in f)
        return pd.read_csv(z.open(archivo),
            sep=None, # autodetecta delimitador
            engine="python",
            encoding="latin1")
```

```python
for anio in range(2023, 2026):
    carpeta = f"{base1}/{anio}/Anual - (Ene-Dic)"
    zip_mod02 = next(
        os.path.join(carpeta, f)
        for f in os.listdir(carpeta)
        if "Modulo02.zip" in f)
    df = leer_enaho(zip_mod02,
        f"Enaho01-{anio}-200")
    print(anio, df.shape)
```

```python
lista = []
for anio in range(2023, 2026):
    carpeta = f"{base1}/{anio}/Anual - (Ene-Dic)"
    zip_mod02 = next(
        os.path.join(carpeta, f)
        for f in os.listdir(carpeta)
        if "Modulo02.zip" in f)
    df = leer_enaho(zip_mod02,
        f"Enaho01-{anio}-200")
    df["anio"] = anio
    lista.append(df)
panel200 = pd.concat(lista, ignore_index=True)
print(panel200.shape)
```
