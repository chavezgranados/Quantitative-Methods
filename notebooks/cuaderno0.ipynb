## **8. Funciones**

```python
def promedio(practica, parcial, final):
    media = (practica * 0.30) + (parcial * 0.30) + (final * 0.40)
    return media
promedio(15, 18, 17)
resultado = promedio(15, 18, 17)

print("Promedio final:", resultado)
```
